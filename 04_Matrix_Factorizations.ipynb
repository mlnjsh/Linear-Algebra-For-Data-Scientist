# 4. Matrix Factorizations / Decompositions

In data science, we rarely "solve" or invert matrices directly in production code. Instead, we use matrix decompositions (breaking a matrix down into a product of simpler matrices). This is analogous to factoring a large number into prime numbers computationally faster and numerically much more stable.

## 4.1 LU Decomposition
Factors a square matrix $A$ into lower triangular matrix $L$ and upper triangular matrix $U$.
$$ A = LU $$

**Why?** Solving $Lx = y$ and then $Uy = b$ (via forward/backward substitution) is much faster than computing $A^{-1}$ when you have to solve $Ax=b$ for many different $b$ vectors.

```python
import numpy as np
import scipy.linalg as la

A = np.array([[4, 3], 
              [6, 3]])

# SciPy provides an easy LU decomposition (returns P, L, U where P is a permutation matrix)
P, L, U = la.lu(A)

print("L (Lower Triangular):\n", L)
print("\nU (Upper Triangular):\n", U)
print("\nReconstructed A (P.dot(L).dot(U)):\n", P.dot(L).dot(U))
```

## 4.2 QR Decomposition
Factors a matrix $A$ into an orthogonal matrix $Q$ ($Q^TQ = I$) and an upper triangular matrix $R$.
$$ A = QR $$

**Why?** QR decomposition is the standard, highly stable algorithm used under the hood to solve Ordinary Least Squares (Linear Regression) and finding eigenvalues.

```python
A_rect = np.array([[1, 2], 
                   [3, 4], 
                   [5, 6]])

Q, R = np.linalg.qr(A_rect)

print("Q (Orthogonal):\n", Q)
print("\nR (Upper Triangular):\n", R)
```

## 4.3 Cholesky Decomposition
A special, ultra-fast decomposition for **symmetric, positive-definite** matrices (like Covariance matrices). It factors $A$ into a lower triangular matrix $L$ and its transpose.
$$ A = LL^T $$

**Why?** Since covariance matrices are always symmetric and positive semi-definite, Cholesky is used heavily in Monte Carlo simulations, Gaussian Processes, and generating correlated random variables.

```python
# A symmetric positive definite matrix
Cov = np.array([[4, 12, -16],
                [12, 37, -43],
                [-16, -43, 98]])

L_chol = np.linalg.cholesky(Cov)
print("Cholesky L:\n", L_chol)
print("\nReconstructed (L @ L.T):\n", L_chol @ L_chol.T)
```

## 4.4 Business & Real-Life Examples

**Business Example 1: Latent Factor Recommender Systems**
E-commerce platforms use matrix factorization on their massive, sparse user-item interaction matrices. By factoring the matrix into smaller "user feature" and "item feature" matrices, they can predict what rating a user *would* give to an item they haven't seen yet.

**Business Example 2: Financial Risk Management**
When banks run Monte Carlo simulations to estimate Value at Risk (VaR) for a portfolio, they use the Cholesky Decomposition of the historical asset covariance matrix to generate thousands of realistically correlated future market scenarios.

**Real-Life Example: Scientific Data Compression**
Climatologists deal with massive grids of weather data (temperature, pressure) over time. Matrix factorization allows them to compress this data by extracting the primary spatial and temporal patterns, making the data small enough to store and analyze.
