# 3. Systems of Linear Equations

A foundational problem in linear algebra is finding the values of unknown variables that satisfy multiple linear equations simultaneously.

## 3.1 Matrix Representation
A system of linear equations:
$$ 2x_1 + 3x_2 = 8 $$
$$ 5x_1 - x_2 = 3 $$

Can be elegantly expressed in matrix-vector form: $\mathbf{Ax = b}$ 

$$ \begin{bmatrix} 2 & 3 \\ 5 & -1 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} 8 \\ 3 \end{bmatrix} $$

```python
import numpy as np

A = np.array([[2, 3], 
              [5, -1]])
b = np.array([8, 3])

# Solving using the inverse: x = A^(-1) b
x_inv = np.linalg.inv(A).dot(b)
print("Solution using Inverse:", x_inv)

# Better, more numerically stable method (Gaussian elimination under the hood)
x_solve = np.linalg.solve(A, b)
print("Solution using linalg.solve:", x_solve)
```

## 3.2 Solving Methods

- **Gaussian Elimination**: Systematically applying row operations to the augmented matrix $[A|b]$ to transform it into upper triangular form, then using back-substitution.
- **LU Decomposition**: Factoring $A$ into a Lower and Upper triangular matrix ($A = LU$) to solve systems faster when we have multiple different $b$ vectors but the same $A$ matrix.
- **Matrix Inversion**: $x = A^{-1}b$. Computationally slow and numerically unstable for large matrices. Do not use in production code!

## 3.3 Underdetermined & Overdetermined Systems

### Overdetermined Systems ($m > n$)
More equations (rows) than unknowns (columns). Usually, there is no exact solution. We look for an approximate solution that minimizes the error ($||Ax - b||_2$). This is the basis of **Ordinary Least Squares (OLS) Linear Regression**.

$ \mathbf{x}_{OLS} = (A^TA)^{-1}A^T b $
(Note: $(A^TA)^{-1}A^T$ is the pseudo-inverse!)

```python
# Overdetermined System Setup (3 points, fitting a 2-parameter line y = mx + c)
A_over = np.array([[1, 1],   # x1 = 1, bias = 1
                   [2, 1],   # x2 = 2, bias = 1
                   [3, 1]])  # x3 = 3, bias = 1
b_over = np.array([1, 2.5, 3.5]) # y labels

# Least squares solution
x_lstsq, residuals, rank, s = np.linalg.lstsq(A_over, b_over, rcond=None)
print("Least Squares Solution (m, c):", x_lstsq)
```

### Underdetermined Systems ($m < n$)
Fewer equations than unknowns. There are infinitely many solutions. This happens when you have more features than samples in your dataset (the "$p > n$ problem"). 

To pick *one* unique solution, we use **Regularization**. For instance, adding an L2 penalty (Ridge Regression) forces the solution to have the smallest possible L2 norm.
