# 10. Additional Important Topics

To round out your linear algebra toolkit for data science, there are a few extra structural concepts and data types you should be familiar with.

## 10.1 The Covariance Matrix
For a dataset $X$ (where rows are samples, columns are features, and data is mean-centered), the Covariance Matrix $\Sigma$ captures the variance of each feature and the correlation between every pair of features.

$$ \Sigma = \frac{1}{n-1} X^T X $$

The diagonal elements are the variances of the individual features. The off-diagonal elements measure how much two features change together.

```python
import numpy as np

# 3 samples, 2 features
X = np.array([[1.2, 2.5],
              [2.4, 4.8],
              [3.1, 6.0]])

# numpy expects rows = features, cols = samples, so we set rowvar=False
Cov_matrix = np.cov(X, rowvar=False)
print("Covariance Matrix:\n", Cov_matrix)
```

## 10.2 Quadratic Forms
A quadratic form is a scalar function of a vector $\mathbf{x}$ generated by a square matrix $A$:
$$ f(\mathbf{x}) = \mathbf{x}^T A \mathbf{x} $$

If $A$ represents a metric (like the inverse covariance matrix in Mahalanobis distance) or an energy landscape, calculating the quadratic form tells you the structural distance or energy of the state vector $\mathbf{x}$. This appears extensively in Support Vector Machine (SVM) dual formulations.

## 10.3 Tensors
While vectors are 1D arrays and matrices are 2D arrays, **Tensors** are $n$-dimensional arrays. 

Data scientists deal with tensors constantly:
- RGB Image $\rightarrow$ 3D Tensor (Height $\times$ Width $\times$ Color Channels).
- Video $\rightarrow$ 4D Tensor (Frames $\times$ Height $\times$ Width $\times$ Color).
- Word batches $\rightarrow$ 3D Tensor (Batch Size $\times$ Sequence Length $\times$ Embedding Dimension).

Libraries like `TensorFlow` and `PyTorch` are essentially highly optimized linear algebra libraries tailored for manipulating these multi-dimensional structures.

## 10.4 Sparse Matrices
In deep learning and NLP, you frequently encounter matrices that are mostly full of zeros (e.g., Bag of Words vectors, TF-IDF matrices, one-hot encodings).

Storing these as dense matrices wastes memory and computational power. We use **Sparse Matrices** (supported in Python by `scipy.sparse`) to store only the non-zero elements and their coordinates.


```python
import scipy.sparse as sp

# Create a dense matrix with mostly zeros
dense_A = np.array([[1, 0, 0, 0],
                    [0, 0, 0, 5],
                    [0, 0, 0, 0],
                    [0, 2, 0, 0]])

# Convert to Compressed Sparse Row (CSR) format
sparse_A = sp.csr_matrix(dense_A)

print("Dense A occupies memory for all 16 elements.")
print("\nSparse representation:\n", sparse_A)
print("\nReconstructing to check:\n", sparse_A.toarray())
```

## 10.5 Business & Real-Life Examples

**Business Example 1: Quantitative Finance (Covariance)**
In Modern Portfolio Theory, financial analysts build a Covariance Matrix of historical stock returns. They use this matrix to construct a portfolio that maximizes expected return while minimizing overall risk, ensuring the assets don't all crash at the exact same time.

**Business Example 2: Large Language Models (Tensors)**
When you type a prompt into ChatGPT, the text is converted into a 3D Tensor (Batch Size $\times$ Sequence Length $\times$ Embedding Dimension). All the self-attention mechanisms and feedforward layers operate natively on these massive tensors.

**Real-Life Example: Medical Imaging (Tensors)**
An MRI or CT scan is not a flat image; it is a 3D volume of the human body. In medical AI, these scans are processed as 3D Tensors (or 4D if it's an fMRI over time) directly by 3D Convolutional Neural Networks to detect anomalies like tumors.
