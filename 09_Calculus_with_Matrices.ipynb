# 9. Calculus with Matrices (Matrix Calculus)

To train machine learning models (especially neural networks), we need to optimize functions (usually minimizing a loss function). Optimization requires finding gradients, and since our parameters are often vectors or matrices, we need *matrix calculus*.

## 9.1 Gradients of Vectors and Matrices
If $f(\mathbf{x})$ is a scalar function taking a vector $\mathbf{x}$ as input, the gradient $\nabla_{\mathbf{x}} f$ is a vector of partial derivatives:

$$ \nabla_{\mathbf{x}} f = \begin{bmatrix} \frac{\partial f}{\partial x_1} \\ \frac{\partial f}{\partial x_2} \\ \vdots \\ \frac{\partial f}{\partial x_n} \end{bmatrix} $$

### Important identities:
- **Gradient of a linear term**: $\nabla_{\mathbf{x}} (\mathbf{w}^T\mathbf{x}) = \mathbf{w}$
- **Gradient of a quadratic form**: $\nabla_{\mathbf{x}} (\mathbf{x}^T A \mathbf{x}) = (A + A^T)\mathbf{x}$. If $A$ is symmetric, this simplifies to $2A\mathbf{x}$.

## 9.2 The Jacobian Matrix
If we have a function $\mathbf{f}(\mathbf{x})$ that takes an $n$-dimensional vector and outputs an $m$-dimensional vector, the derivative is an $m \times n$ matrix called the **Jacobian** ($J$). 

The element $J_{i,j} = \frac{\partial f_i}{\partial x_j}$. The Jacobian plays a central role in the generalized Chain Rule used in backpropagation.

## 9.3 The Hessian Matrix
If $f(\mathbf{x})$ is a scalar function, taking the derivative twice gives us a matrix of second-order partial derivatives called the **Hessian** ($H$).

$H_{i,j} = \frac{\partial^2 f}{\partial x_i \partial x_j}$

The Hessian is always symmetric. Its eigenvalues tell us about the curvature of the loss landscape:
- All positive eigenvalues $\rightarrow$ We are at a **local minimum**.
- All negative eigenvalues $\rightarrow$ We are at a **local maximum**.
- Mixed eigenvalues $\rightarrow$ We are at a **saddle point**.

```python
import autograd.numpy as np
from autograd import grad, hessian

# Let's define a simple quadratic function: f(x, y) = 3x^2 + 5y^2 + 2xy
# The minimum is clearly at (0, 0)
def f(w):
    x, y = w[0], w[1]
    return 3 * x**2 + 5 * y**2 + 2 * x * y

# Initial point
w_init = np.array([1.0, 2.0])

gradient_func = grad(f)
hessian_func = hessian(f)

print("Function value at w_init:", f(w_init))
print("\nGradient at w_init:\n", gradient_func(w_init))
print("\nHessian Matrix (constant for quadratics):\n", hessian_func(w_init))
```

## 9.4 Applications in Data Science

- **Gradient Descent**: The core optimization algorithm in deep learning ($W_{new} = W_{old} - \alpha \nabla_W L$).
- **Backpropagation**: An application of the multivariate chain rule, requiring continuous multiplication of Jacobian matrices from layer to layer.
- **Newton's Method**: An advanced optimization algorithm that uses the inverse of the Hessian matrix ($H^{-1}$) to find minima much faster, though computing $H^{-1}$ is computationally prohibitive for deep neural networks.
