# 1. Vectors

Welcome to the first module of **Linear Algebra for Data Scientists**. Vectors form the foundation of linear algebra. In machine learning, vectors represent data points, weights in a neural network, or text embeddings.

## 1.1 Definition & Notation
A vector is an ordered tuple (or list) of numbers. If a vector $v$ has $n$ elements, we say it belongs to an $n$-dimensional space, denoted as $\mathbb{R}^n$.

For example, a vector representing the properties of a house (price, square footage, number of bedrooms) could be written as:
$$ \mathbf{x} = \begin{bmatrix} 500000 \\ 2500 \\ 3 \end{bmatrix} $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Defining a vector in NumPy
v = np.array([500000, 2500, 3])
print("Vector v:", v)
print("Shape of v:", v.shape)
```

## 1.2 Basic Operations

### Vector Addition
Vectors are added element-wise: $\mathbf{a} + \mathbf{b} = \begin{bmatrix} a_1 + b_1 \\ a_2 + b_2 \\ ... \\ a_n + b_n \end{bmatrix}$

### Scalar Multiplication
A vector is multiplied by a scalar (a single number) element-wise: $c \mathbf{a} = \begin{bmatrix} c a_1 \\ c a_2 \\ ... \\ c a_n \end{bmatrix}$

### Dot Product (Inner Product)
The dot product of two vectors produces a scalar value. It measures how much two vectors point in the same direction.
$$ \mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^{n} a_i b_i = a_1b_1 + a_2b_2 + \dots + a_nb_n $$

If the dot product is zero, the vectors are **orthogonal** (perpendicular perpendicular).

### Cross Product
Defined only for 3D vectors, the cross product produces a new vector that is orthogonal to both input vectors.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Vector Addition
print("a + b =", a + b)

# Scalar Multiplication
print("3 * a =", 3 * a)

# Dot Product
print("Dot product (a . b) =", np.dot(a, b))

# Cross Product
print("Cross product (a x b) =", np.cross(a, b))
```

## 1.3 Norms (Vector Magnitude)
A norm is a function that assigns a strictly positive length or size to each vector. It is often used as a distance measure or a loss/regularization term in machine learning.

- **L1 Norm (Manhattan Distance)**: Sum of absolute values. Promotes sparsity (Lasso regression).
$$ \|\mathbf{v}\|_1 = \sum |v_i| $$

- **L2 Norm (Euclidean Distance)**: The standard "straight line" distance. (Ridge regression).
$$ \|\mathbf{v}\|_2 = \sqrt{\sum v_i^2} $$


```python
v = np.array([3, -4])

# L1 Norm
l1_norm = np.linalg.norm(v, ord=1)
print("L1 Norm:", l1_norm)

# L2 Norm
l2_norm = np.linalg.norm(v, ord=2)
print("L2 Norm:", l2_norm)
```

## 1.4 Linear Combinations & Span
A **linear combination** of a set of vectors $\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_k$ is formed by multiplying each vector by a scalar and adding the results:
$$ \mathbf{u} = c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \dots + c_k\mathbf{v}_k $$

The **span** of a set of vectors is the set of all possible linear combinations of those vectors. If the span of vectors covers the entire space (e.g., $\mathbb{R}^3$), those vectors can "build" any point in that space.

## 1.5 Linear Independence & Basis
A set of vectors is **linearly dependent** if at least one vector in the set can be written as a linear combination of the others. This means it is redundant; it doesn't add any "new dimension" to the span.

A set of vectors is **linearly independent** if no vector can be written as a linear combination of the others. 

A **basis** for a vector space is a set of linearly independent vectors that span the entire space. The number of vectors in a basis is the **dimension** of the space.

In data science, we usually want features that are linearly independent. Linearly dependent features cause issues with multicollinearity, making models unstable.

## 1.6 Applications in Data Science
- **Representing observations**: A single row in a pandas DataFrame is a feature vector.
- **Word Embeddings**: In NLP (e.g., Word2Vec, BERT), words are represented as dense vectors in high-dimensional space.
- **Feature Engineering**: Creating new features by taking linear combinations of existing ones.
