# 2. Matrices

A matrix is a rectangular array of numbers arranged in rows and columns. In data science, matrices are everywhere: a dataset is a matrix where rows are samples and columns are features; an image is a matrix of pixel intensities; weights in a neural network layer form a matrix.

## 2.1 Definition & Notation
An $m \times n$ matrix $A$ has $m$ rows and $n$ columns. The element in the $i$-th row and $j$-th column is denoted as $A_{i,j}$.

```python
import numpy as np

A = np.array([[1, 2, 3], 
              [4, 5, 6]])
print("Matrix A:\n", A)
print("Shape (rows, columns):", A.shape)
```

## 2.2 Basic Operations

### Matrix Addition
Matrices of the same shape can be added element-wise: $C_{i,j} = A_{i,j} + B_{i,j}$

### Matrix Transpose ($A^T$)
The transpose of a matrix flips it over its diagonal, switching the row and column indices: $(A^T)_{i,j} = A_{j,i}$

### Matrix Multiplication
To multiply matrix $A$ (shape $m \times n$) by matrix $B$ (shape $n \times p$), the number of columns in $A$ must equal the number of rows in $B$. The resulting matrix $C$ has shape $m \times p$. The element $C_{i,j}$ is the dot product of the $i$-th row of $A$ and the $j$-th column of $B$.
$$ C = AB $$

```python
B = np.array([[7, 8], 
              [9, 10], 
              [11, 12]])

# Transpose
print("A Transpose:\n", A.T)

# Matrix Multiplication (Dot Product of matrices)
C = np.dot(A, B) # equivalent to A @ B in modern Python
print("\nA x B =\n", C)
```

## 2.3 Special Matrices

- **Identity Matrix ($I$)**: A square matrix with 1s on the main diagonal and 0s elsewhere. $AI = IA = A$.
- **Diagonal Matrix**: A square matrix where all off-diagonal elements are zero.
- **Symmetric Matrix**: A square matrix that is equal to its transpose ($A = A^T$). Very important in PCA (covariance matrices).
- **Orthogonal Matrix**: A square matrix whose rows and columns are orthogonal unit vectors. $A^TA = AA^T = I$, so $A^T = A^{-1}$.
- **Positive Definite Matrix**: A symmetric matrix $A$ where $\mathbf{x}^T A \mathbf{x} > 0$ for all non-zero vectors $\mathbf{x}$. Crucial for optimization (convexity).

## 2.4 Matrix Rank
The **rank** of a matrix is the maximum number of linearly independent column vectors (which magically equals the number of linearly independent row vectors). 
- If rank is less than the number of columns/rows, the matrix is "rank-deficient".
- In machine learning, a full-rank matrix means no features are perfectly redundant.

```python
# A rank-deficient matrix (row 2 is a multiple of row 1)
M = np.array([[1, 2, 3], 
              [2, 4, 6], 
              [7, 8, 9]])
rank = np.linalg.matrix_rank(M)
print("Matrix Rank:", rank)
```

## 2.5 Determinant 
The determinant ($det(A)$ or $|A|$) is a scalar computed from a square matrix. 
Geometrically, it represents the scaling factor of the linear transformation described by the matrix. If a shape transforms by matrix $A$, its area/volume scales by $|det(A)|$.
If $det(A) = 0$, the matrix is **singular** (not invertible).

```python
S = np.array([[4, 2], [1, 3]])
det = np.linalg.det(S)
print("Determinant of S:", det)
```

## 2.6 Inverse & Pseudo-inverse
The **inverse** of a square matrix $A$ is denoted $A^{-1}$, where $AA^{-1} = I$.
If $A$ is not square or is singular, it cannot be inverted. However, we can compute the **Moore-Penrose pseudo-inverse** ($A^+$), which provides the best possible least-squares solution to a system.

```python
# Inverse
S_inv = np.linalg.inv(S)
print("Inverse of S:\n", S_inv)

# Pseudo-inverse of the non-square matrix A
A_pinv = np.linalg.pinv(A)
print("\nPseudo-inverse of A:\n", A_pinv)
```
