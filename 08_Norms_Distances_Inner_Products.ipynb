# 8. Norms, Distances, and Inner Products

Measuring "how big" something is or "how far apart" two things are is the core of most machine learning loss functions and clustering algorithms. 

## 8.1 Vector Norms
A norm $||\mathbf{x}||$ measures the length or magnitude of a vector.

- **L0 Norm**: Not rigorously a norm, it simply counts the number of non-zero elements. Used to measure strict sparsity.
- **L1 Norm (Manhattan)**: $||\mathbf{x}||_1 = \sum |x_i|$. Important for Lasso regression because it drives some weights exactly to zero.
- **L2 Norm (Euclidean)**: $||\mathbf{x}||_2 = \sqrt{\sum x_i^2}$. The standard notion of distance. Important for Ridge regression.
- **L-Infinity Norm (Max)**: $||\mathbf{x}||_\infty = \max |x_i|$. The absolute value of the largest component.

```python
import numpy as np

v = np.array([3, -4, 0, 1.5])

print("L0 Norm (non-zeros):", np.count_nonzero(v))
print("L1 Norm:", np.linalg.norm(v, ord=1))
print("L2 Norm:", np.linalg.norm(v, ord=2))
print("L-Infinity Norm:", np.linalg.norm(v, ord=np.inf))
```

## 8.2 Matrix Norms
We can also measure the "size" of a matrix. This is crucial for bounding errors and regularization in deep learning.

- **Frobenius Norm**: $||A||_F = \sqrt{\sum_i \sum_j A_{i,j}^2}$. This is just the L2 norm of the matrix if you flattened it into a long vector.
- **Spectral Norm**: (Also called the L2 Matrix Norm) The maximum singular value of the matrix. It represents the maximum amount the matrix can stretch a vector.

```python
A = np.array([[1, 2], [3, 4]])

print("Frobenius Norm:", np.linalg.norm(A, ord='fro'))
print("Spectral Norm (Max singular value):", np.linalg.norm(A, ord=2))
```

## 8.3 Distances and Inner Products

### Distance
Once you define a norm, you automatically define a distance metric between two vectors: $d(\mathbf{u}, \mathbf{v}) = ||\mathbf{u} - \mathbf{v}||$.

### Inner Products
The standard inner product in $\mathbb{R}^n$ is the dot product. It relates geometry and algebra: $\mathbf{u} \cdot \mathbf{v} = ||\mathbf{u}||_2 ||\mathbf{v}||_2 \cos(\theta)$

- **Cosine Similarity**: Isolates just the angle $\theta$ between vectors, ignoring their magnitudes. Extremely common in Natural Language Processing for comparing document embeddings.
$$ \text{Cosine Similarity} = \cos(\theta) = \frac{\mathbf{u} \cdot \mathbf{v}}{||\mathbf{u}||_2 ||\mathbf{v}||_2} $$

```python
u = np.array([1, 0, -1])
v = np.array([-1, 1, 0])

cosine_sim = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
print("Cosine Similarity:", cosine_sim)
```

## 8.4 Applications in Data Science
- **Regularization Terms**: Penalizing model complexity (L1/L2 weights) to prevent overfitting.
- **Loss Functions**: Mean Squared Error (MSE) is the squared L2 norm of the error vector.
- **Clustering algorithms**: K-Means relies heavily on the Euclidean (L2) distance metric, while text-clustering might use Cosine Similarity.

## 8.5 Business & Real-Life Examples

**Business Example 1: Fraud Detection Systems**
Banks map historical credit card transactions into a vector space. When a new transaction occurs, they calculate the Euclidean distance (L2 Norm) between it and the user's normal cluster of behavior. If the distance is too large, it is flagged as anomalous.

**Business Example 2: Recommendation Engines (Similarity)**
When you read an article on a news site, the site wants to recommend similar articles. It converts all articles into vectors and calculates the Cosine Similarity (Inner Product) between them, recommending the ones with the highest similarity score.

**Real-Life Example: GPS and Routing**
If you use a drone delivery service, the energy cost is proportional to the standard "Euclidean" (L2) distance. But if you are a delivery driver navigating a city grid (like Manhattan), the time and fuel cost is proportional to the "Manhattan" (L1) distance.
