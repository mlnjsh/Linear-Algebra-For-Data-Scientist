# 5. Eigenvalues and Eigenvectors

When a square matrix $A$ multiplies a vector $\mathbf{v}$, it typically rotates and stretches the vector. However, for certain special vectors called **eigenvectors**, the matrix only stretches or squishes them without changing their direction.

## 5.1 Definition & Computation
$$ A\mathbf{v} = \lambda\mathbf{v} $$

- $\mathbf{v}$ is the eigenvector.
- $\lambda$ (lambda) is the scalar eigenvalue (the amount of scaling).

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[2, 1], 
              [1, 2]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:", eigenvalues)
# Note: eigenvectors are returned as COLUMNS of the 2D array
print("\nEigenvectors:\n", eigenvectors)
```

## 5.2 Properties
- **Trace**: The sum of diagonal elements of $A$ equals the sum of its eigenvalues.
- **Determinant**: The determinant of $A$ equals the product of its eigenvalues.
- If a matrix is symmetric (like Covariance or Graph Laplacian matrices), all its eigenvalues are real, and its eigenvectors are orthogonal to each other.

## 5.3 Eigendecomposition (Diagonalization)
If a square $n \times n$ matrix $A$ has $n$ linearly independent eigenvectors, it can be factored into:
$$ A = V \Lambda V^{-1} $$
Where $V$ is a matrix whose columns are the eigenvectors, and $\Lambda$ (Lambda) is a diagonal matrix of the eigenvalues.

This makes matrix powers incredibly easy to compute: $A^k = V \Lambda^k V^{-1}$, a critical property for Markov Chains and PageRank.

```python
V = eigenvectors
Lambda = np.diag(eigenvalues)
V_inv = np.linalg.inv(V)

A_reconstructed = V.dot(Lambda).dot(V_inv)
print("Reconstructed A:\n", A_reconstructed)

# Notice that since A was symmetric, V is orthogonal, meaning V_inv is just V.T!
```

## 5.4 Applications in Data Science

- **Principal Component Analysis (PCA)**: The eigenvectors of the data's covariance matrix represent the "principal components" (directions of maximum variance). The eigenvalues tell you how much variance is explained by each component.
- **Spectral Clustering**: Grouping data points based on the eigenvectors of the graph Laplacian matrix.
- **Google PageRank**: The steady-state probability of landing on a webpage is exactly the eigenvector corresponding to the eigenvalue $\lambda=1$ of the network's transition matrix.

## 5.5 Business & Real-Life Examples

**Business Example 1: Google PageRank**
The original algorithm that made Google the dominant search engine is based entirely on eigenvectors. The internet is modeled as a massive transition matrix of links. The "importance" of every webpage is simply the eigenvector corresponding to the eigenvalue of 1 for this matrix.

**Business Example 2: Market Risk and PCA**
Hedge funds use PCA on the covariance matrix of thousands of stocks. The resulting eigenvectors (principal components) usually correspond to hidden market factors (like "the overall tech sector" or "interest rate sensitivity"). The eigenvalues tell them which factor drives the most risk.

**Real-Life Example: Structural Engineering**
When designing skyscrapers or bridges, engineers build matrices representing the physical stiffness and mass of the structure. The eigenvalues of these matrices tell them the natural resonant frequencies of the building, which is critical for making them earthquake and wind-proof.
