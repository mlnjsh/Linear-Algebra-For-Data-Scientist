# 7. Linear Transformations

Every matrix can be thought of as a function or an action. If you take a vector $\mathbf{x}$ and multiply it by a matrix $A$, you transform $\mathbf{x}$ into a new vector $\mathbf{y}$.
$$ \mathbf{y} = A\mathbf{x} $$

A transformation is *linear* if lines stay lines (they don't bend) and the origin remains fixed at $(0,0)$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Let's define a square shape (four points as column vectors)
square = np.array([[0, 1, 1, 0, 0], 
                   [0, 0, 1, 1, 0]])

# Define a Transformation Matrix (Shear) 
T_shear = np.array([[1, 1], 
                    [0, 1]])

# Apply transformation
square_transformed = T_shear.dot(square)

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.plot(square[0, :], square[1, :], marker='o')
plt.title("Original Object")
plt.xlim(-1, 3); plt.ylim(-1, 3)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(square_transformed[0, :], square_transformed[1, :], marker='o', color='red')
plt.title("After Matrix Subtraction (Shear)")
plt.xlim(-1, 3); plt.ylim(-1, 3)
plt.grid(True)
plt.show()
```

## 7.1 Change of Basis
When we represent data points (vectors), we represent them using our standard "basis" (the standard x, y, z axes). However, another perspective (basis) might make patterns in the data much clearer.
Multiplying data by a transformation matrix is essentially shifting how you view that data's coordinate system.

## 7.2 Kernel (Null Space) & Image (Column Space)
- **Kernel ($Null(A)$)**: The set of all vectors $\mathbf{x}$ that the matrix crushes down to zero ($A\mathbf{x} = \mathbf{0}$). In ML, if a weight matrix has a large null space, it means the model is ignoring a lot of input features.
- **Image ($Col(A)$)**: The set of all possible outputs $A\mathbf{x}$. It is the span of the columns of $A$. If the column space doesn't span the whole output dimension, your model cannot possibly output certain combinations of predictions.

## 7.3 Applications in Data Science

- **Neural Network Layers**: The core of a dense layer in a neural network is an affine transformation: $\mathbf{y} = \sigma(W\mathbf{x} + \mathbf{b})$. The $W$ matrix is a linear transformation mapping features from the input dimension to a new hidden dimension.
- **Data Augmentation**: In computer vision, images are represented as matrices. Rotating, scaling, or shearing images to create more training data is done strictly via matrix multiplications.
- **Coordinate Transformations**: Mapping data into higher dimensions (feature maps/kernel trick in SVMs) or lower dimensions (PCA).

## 7.4 Business & Real-Life Examples

**Business Example 1: Image Data Augmentation**
Companies building computer vision models for self-driving cars need massive amounts of varied training data. They use linear transformations (rotation, scaling, shearing matrices) to automatically generate thousands of new variation images from a single original photograph.

**Business Example 2: Automated Translation Models**
In machine translation, words are mapped into high-dimensional vector spaces. A linear transformation matrix can often be learned to map the English word space directly onto the Spanish word space, effectively translating words purely mathematically.

**Real-Life Example: 3D Graphics and Video Games**
Every time a character moves in a video game, their 3D coordinates are continuously multiplied by linear transformation matrices. A "Projection Matrix" then flattens those 3D coordinates into the 2D plane of your computer monitor.
