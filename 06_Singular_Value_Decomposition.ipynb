# 6. Singular Value Decomposition (SVD)

SVD is the "crown jewel" of linear algebra. While eigendecomposition only works for square (and usually diagonalizable) matrices, **SVD works for ANY matrix**, no matter its shape.

## 6.1 Definition
Any $m \times n$ matrix $A$ can be factorized into:
$$ A = U \Sigma V^T $$

- $U$ ($m \times m$): Orthogonal matrix representing left singular vectors (outputs/columns space basis).
- $\Sigma$ ($m \times n$): Diagonal matrix of singular values in descending order (scaling factors, related to eigenvalues of $AA^T$).
- $V^T$ ($n \times n$): Orthogonal matrix representing right singular vectors (inputs/rows space basis).

```python
import numpy as np

# An arbitrary rectangular matrix (3x2)
A = np.array([[1, 2], 
              [3, 4], 
              [5, 6]])

U, S, VT = np.linalg.svd(A, full_matrices=False) # Reduced SVD

print("U (Left Singular Vectors):\n", U)
print("\nSigma (Singular Values):", S)
print("\nV^T (Right Singular Vectors transposed):\n", VT)

# Reconstructing A
A_reconstructed = U.dot(np.diag(S)).dot(VT)
print("\nReconstructed A:\n", A_reconstructed)
```

## 6.2 Low-Rank Approximation (Eckart–Young Theorem)
If you want to compress a matrix (e.g., an image) by dropping less important information, you can simply keep only the top $k$ singular values and their corresponding vectors from $U$ and $V^T$.

According to the Eckart-Young theorem, $A_k = U_k \Sigma_k V_k^T$ is the mathematically proven *best* rank-$k$ approximation of the original matrix $A$ (in terms of Frobenius norm error).

## 6.3 Applications in Data Science

- **Dimensionality Reduction (PCA)**: Under the hood, modern implementations of PCA (like in `scikit-learn`) actually compute the SVD of the data matrix rather than the eigendecomposition of the covariance matrix because SVD is much more numerically stable.
- **Recommender Systems**: Algorithms like Latent Factor Models in Collaborative Filtering factorize the user-item rating matrix to predict missing ratings (e.g., Netflix Prize).
- **Topic Modeling**: Latent Semantic Analysis (LSA) in NLP uses SVD on the Document-Term matrix to find hidden topics.
- **Computing the Pseudo-Inverse**: SVD is how numpy computers $A^+$ securely: $A^+ = V \Sigma^+ U^T$.

## 6.4 Business & Real-Life Examples

**Business Example 1: Latent Semantic Analysis (NLP)**
In natural language processing, companies use SVD on a "document-term" matrix (where rows are words and columns are documents). This reduces the data down to a smaller number of "concept" dimensions, allowing search engines to realize that a search for "car" should also return documents containing "automobile".

**Business Example 2: Customer Churn Prediction**
Telecomm companies have massive datasets with hundreds of columns about user behavior. Before feeding this into a machine learning model, they often use SVD (via PCA) to remove noise and reduce the hundreds of columns down to the 20 most important abstract features, preventing overfitting.

**Real-Life Example: Image Compression**
When transmitting images from the Mars Rover or satellites back to Earth, bandwidth is highly limited. By taking the SVD of the image and keeping only the top 10% of the singular values, you can transmit a highly accurate approximation of the image using a fraction of the data.
